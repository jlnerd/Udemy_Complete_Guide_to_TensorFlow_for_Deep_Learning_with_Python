```python
import numpy as np
import sklearn
from sklearn import *
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create Some Data

```python
data = sklearn.datasets.make_blobs(n_samples=100, 
                  n_features=3,
                  centers=2,
                  random_state=101)
```

## Scale the Data

```python
scaler = sklearn.preprocessing.MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(data[0])
```

```python
data_x = scaled_data[:,0]
data_y = scaled_data[:,1]
data_z = scaled_data[:,2]
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(data_x,data_y,data_z,c=data[1])
plt.show()
```

# The Linear Autoencoder

```python
import tensorflow as tf
```

```python
n_inputs = 3
n_HLs = 2
n_outputs = n_inputs

learning_rate = 0.01
```

### Placeholder

Notice there is no real label here, just X.

```python
x = tf.placeholder(tf.float32, shape = [None, n_inputs])
```

### Layers

Using the fully_connected layers API, we **do not** provide an activation function!

```python
HL = tf.contrib.layers.fully_connected(x, n_HLs, activation_fn=None)
outputs = tf.contrib.layers.fully_connected(HL, n_outputs, activation_fn=None)
```

### Loss Function

```python
loss = tf.reduce_mean(tf.square(outputs - x))
```

### Optimizer

```python
optimizer = tf.train.AdamOptimizer(learning_rate)
minimizer = optimizer.minimize(loss)
```

### Init

```python
init = tf.global_variables_initializer()
```

## Running the Session

```python
n_steps = 1000
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(n_steps):
        sess.run(minimizer,feed_dict={x:scaled_data})
    output_2d = HL.eval(feed_dict={x:scaled_data})
```

```python
plt.plot(output_2d[:,0],output_2d[:,1],'o')
```
